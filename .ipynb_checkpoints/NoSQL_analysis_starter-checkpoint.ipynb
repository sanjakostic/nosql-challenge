### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    'LocalAuthorityName': 'London',
    'RatingValue': {'$gte': 4}
}

# Use count_documents to display the number of documents in the result
num_documents = establishments.count_documents(query)
print("Number of documents:", num_documents)

# Display the first document in the results using pprint
first_document = establishments.find_one(query)
print("First document:")
pprint(first_document)

# Convert the result to a Pandas DataFrame
result_cursor = establishments.find(query)
df = pd.DataFrame(list(result_cursor))

# Display the number of rows in the DataFrame
print("Number of rows in DataFrame:", len(df))

# Display the first 10 rows of the DataFrame
print("First 10 rows of DataFrame:")
df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Create a dictionary for the new restaurant data
new_rest = {
    "BusinessName": "Penang Flavours",
    "BusinessType": "Restaurant/Cafe/Canteen",
    "BusinessTypeID": "",
    "AddressLine1": "Penang Flavours",
    "AddressLine2": "146A Plumstead Rd",
    "AddressLine3": "London",
    "AddressLine4": "",
    "PostCode": "SE18 7DY",
    "Phone": "",
    "LocalAuthorityCode": "511",
    "LocalAuthorityName": "Greenwich",
    "LocalAuthorityWebSite": "http://www.royalgreenwich.gov.uk",
    "LocalAuthorityEmailAddress": "health@royalgreenwich.gov.uk",
    "scores": {
        "Hygiene": "",
        "Structural": "",
        "ConfidenceInManagement": ""
    },
    "SchemeType": "FHRS",
    "geocode": {
        "longitude": "0.08384000",
        "latitude": "51.49014200"
    },
    "RightToReply": "",
    "Distance": 4623.9723280747176,
    "NewRatingPending": True
}

# Insert the new restaurant into the collection
establishments.insert_one(new_rest)

```

```python
penang_flavours = establishments.find_one({'BusinessName': 'Penang Flavours'})

if penang_flavours is not None:
    geocode = penang_flavours.get('geocode', {})
    latitude_str = geocode.get('latitude')
    longitude_str = geocode.get('longitude')

    try:
        latitude = float(latitude_str)
        longitude = float(longitude_str)
    except (TypeError, ValueError):
        latitude = None
        longitude = None

    if latitude is not None and longitude is not None:
        degree_search = 0.01

        query = {
            'RatingValue': 5,
            'geocode.latitude': {'$gte': latitude - degree_search, '$lte': latitude + degree_search},
            'geocode.longitude': {'$gte': longitude - degree_search, '$lte': longitude + degree_search}
        }

        sort = [('scores.Hygiene', 1)]  # Sort by hygiene score in ascending order

        limit = 5

        result_cursor = establishments.find(query).sort(sort).limit(limit)

        num_documents = establishments.count_documents(query)
        print("Number of documents:", num_documents)

        first_document = result_cursor.next()
        print("First document:")
        pprint(first_document)

        result_cursor = establishments.find(query).sort(sort).limit(limit)
        df = pd.DataFrame(list(result_cursor))

        print("Number of rows in DataFrame:", len(df))

        print("First 10 rows of DataFrame:")
        df.head(10)
    else:
        print("Latitude or longitude not available for 'Penang Flavours'.")
else:
    print("No document found with BusinessName 'Penang Flavours'.")
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
pipeline = [
    # Match establishments with a hygiene score of 0
    {"$match": {"scores.Hygiene": 0}},
    # Group the matches by Local Authority
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    # Sort the matches from highest to lowest
    {"$sort": {"count": -1}}
]
result = establishments.aggregate(pipeline)

num_documents = len(list(result))
print("Number of documents:", num_documents)

result = establishments.aggregate(pipeline)
print("First 10 results:")
for i, doc in enumerate(result):
    if i >= 10:
        break
    pprint(doc)
```

```python
result = list(establishments.aggregate(pipeline))

df = pd.DataFrame(result)

num_rows = len(df)
print("Number of rows in DataFrame:", num_rows)

print("First 10 rows of DataFrame:")
df.head(10)
```
